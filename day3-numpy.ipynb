```python
# 64. ２つのベクトルa,bがあって、aベクトルに、bで与えられたインデックスに箇所に1を加える
# 自分の回答
a = np.ones(10)
b = np.random.randint(0,10,5)
print("64:")
print(a,b)
for el in b:
    a[el] += 1
print(a)
# 模範解答
# 1. bitcountを使う方法
a = np.ones(10)
a += np.bincount(b, minlength=len(a))
print(a)
# 2. np.add.atを使う
a = np.ones(10)
np.add.at(a,b,1)
print(a)

# 65. aベクトルの値を用いて新しいベクトルを、bのインデックスに従って作る
print("65:")
a = np.arange(10)
b = np.random.randint(0,10,10)
print(a,b)
print(np.bincount(b,a))

# 66. (dtype=ubyte)の(w,h,3)画像全体を定義し、uniqueな値の数計算する(色はRGB256階調)
# np.unique(F)は、uniqueな値の配列を返す(setみたいなもの)
print("66:")
w,h = 16,16
I = np.random.randint(0,2,(h,w,3)).astype(np.ubyte)
# print(I[:,:,0])とprint(I[...,0])は同義

F = I[...,0]*(256*256) + I[...,1]*256 +I[...,2]
print(np.unique(F))
n = len(np.unique(F))
print(n)

# 67, 4次元配列において、最後の2軸だけ一気に足す方法は？
print("67:")
A = np.random.randint(0,5,(2,2,2,2))
print(A)
B = A.sum(axis=(-1,-2))
print(B)
# 別解
print(A.reshape(A.shape[:-2]+(-1,)).sum(axis=-1))

# 68. 1次元ベクトルaを用意する。aと同じサイズで用意されたインデックスベクトル(min:0, max:9) s に従って構成されるsubsetベクトルの平均を求めよ
print("68:")
# 自分の回答
a = np.random.random(100)
s = np.random.randint(0,10,100)
c = np.zeros(10)
for (v,el) in zip(a,s):
    c[el] += v
print(c/np.bincount(s))
# 模範解答(pandas使う)
import pandas as pd
print(pd.Series(a).groupby(s).mean())
# bitcount等
D_sums = np.bincount(s, weights=a)
D_counts = np.bincount(s)
D_means = D_sums / D_counts
print(D_means)

"""
bincount便利
axisは何個か指定できる
np.uniqueもなかなか便利
"""
```

```python
# 69. 内積の対角成分だけを取り出す
print("69:")
A = np.random.randint(0,5,(3,3))
B = np.random.randint(0,5,(3,3))
print(np.diag(np.dot(A,B)))
# もっと早いver.(うまく計算を利用)
print(np.sum(A*B.T, axis=1))
# もっともっと早いver.(einsum)
print(np.einsum("ij,ji->i",A,B))
# 70. [1,2,3,4,5]に対し、各値の間に3つの連続したゼロが挿入された新しいベクトルを作成する方法は？
a = np.array([1,2,3,4,5])
nz = 3
a0 = np.zeros((len(a)-1)*nz+len(a))
a0[::nz+1] = a
print(a0)
```

```python
# 73, 三角形を記述する10個のtripletsがある。全ての三角形を作る線分の中で、uniqueなモノを書き出す。
print("73:")
faces = np.random.randint(0,100,(10,3))
print(faces)
F = np.roll(faces.repeat(2,axis=1),-1,axis=1)
print(F)
F = F.reshape(len(F)*3,2)
F = np.sort(F,axis=1)
print(F)
# https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.ndarray.view.html
G = F.view(dtype=[('p0',F.dtype),('p1',F.dtype)] )
print(G)
G = np.unique(G)
print(G)
```

```python
# 74, bincountされた配列Cに対して、np.bincount(A)==CとなるようなAをどう作るか
# np.repeat : https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.repeat.html: 
print("74:")
C = np.bincount([1,2,3,3,2,4])
print(C)
A = np.repeat(np.arange(len(C)),C)
print(A)

# 75. array上でsliding windowを使った平均算出法
# pandasで良くない？
# 自分の回答
print("75:")
a = np.arange(30)
window = 10
b = np.zeros(len(a))
for i in range(len(a)):
    if i == 0:
        b[i] = a[i]
    elif i < 10:
        b[i] = a[:i+1].mean()
    else:
        b[i] = a[i-10+1:i+1].mean()
print(b)
# 模範解答 : cumsum
# cumsum https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.cumsum.html
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
print(moving_average(a,window))

# 76. 一次元ベクトルaに対して一行目が(a[0],a[1],a[2])となって、以下の行はインデックスを1ずつずらした値が入るような行列をつくるう
# numpy.lib.stride_tricksの利用
# https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.lib.stride_tricks.as_strided.html
print("76:")
a = np.arange(10)
def rolling(a, window):
    shape = (a.size-window+1,window)
    strides = (a.itemsize, a.itemsize)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
A = rolling(a,5)
print(A)

# 77. booleanをひっくり返す、floatの記号をひっくり返す
# logical_not, negativeという関数
print("77:")
a = np.random.randint(0,2,10)
print(a)
np.logical_not(a, out=a)
print(a)
a = np.random.uniform(-1.0,1.0,10)
print(a)
np.negative(a, out=a)
print(a)
```
