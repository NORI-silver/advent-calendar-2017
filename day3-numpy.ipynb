```python
# 74, bincountされた配列Cに対して、np.bincount(A)==CとなるようなAをどう作るか
# np.repeat : https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.repeat.html: 
print("74:")
C = np.bincount([1,2,3,3,2,4])
print(C)
A = np.repeat(np.arange(len(C)),C)
print(A)

# 75. array上でsliding windowを使った平均算出法
# pandasで良くない？
# 自分の回答
print("75:")
a = np.arange(30)
window = 10
b = np.zeros(len(a))
for i in range(len(a)):
    if i == 0:
        b[i] = a[i]
    elif i < 10:
        b[i] = a[:i+1].mean()
    else:
        b[i] = a[i-10+1:i+1].mean()
print(b)
# 模範解答 : cumsum
# cumsum https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.cumsum.html
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
print(moving_average(a,window))

# 76. 一次元ベクトルaに対して一行目が(a[0],a[1],a[2])となって、以下の行はインデックスを1ずつずらした値が入るような行列をつくるう
# numpy.lib.stride_tricksの利用
# https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.lib.stride_tricks.as_strided.html
print("76:")
a = np.arange(10)
def rolling(a, window):
    shape = (a.size-window+1,window)
    strides = (a.itemsize, a.itemsize)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
A = rolling(a,5)
print(A)

# 77. booleanをひっくり返す、floatの記号をひっくり返す
# logical_not, negativeという関数
print("77:")
a = np.random.randint(0,2,10)
print(a)
np.logical_not(a, out=a)
print(a)
a = np.random.uniform(-1.0,1.0,10)
print(a)
np.negative(a, out=a)
print(a)

# 78. ２つの点p0,p1がある。ある点pにおいて、p0,p1が作る直線と点pの距離を求めよ。
print("78:")
# 自分の回答
def distance(p0,p1,p):
    d0 = p0-p
    d1 = p1-p
    d = p0-p1
    #三角形の面積
    s = np.sqrt((d0**2).sum()*(d1**2).sum()-np.dot(d0,d1)**2)/2
    length = np.sqrt(d[0]**2+d[1]**2)
    if length == 0:
        return("error:div#0")
    return 2*s/length
p0 = np.array([0,0])
p1 = np.array([3,3])
p = np.array([3,0])
print(distance(p0,p1,p))
# 模範解答
def distance(P0, P1, p):
    T = P1 - P0
    L = (T**2).sum(axis=1) #こういうの出来るのね。
    U = -((P0[:,0]-p[...,0])*T[:,0] + (P0[:,1]-p[...,1])*T[:,1]) / L
    U = U.reshape(len(U),1)
    D = P0 + U*T - p
    return np.sqrt((D**2).sum(axis=1))

P0 = np.random.uniform(-10,10,(10,2))
P1 = np.random.uniform(-10,10,(10,2))
p  = np.random.uniform(-10,10,( 1,2))
print(distance(P0, P1, p))

# 79. 点群の集合P0,P1,Pがあって、Pの各要素の点からP0,P1が作る直線の距離を計算して配列にせよ。
# リスト内包表記を使って、np.array()でまとめる
P0 = np.random.uniform(-10, 10, (10,2))
P1 = np.random.uniform(-10,10,(10,2))
p = np.random.uniform(-10, 10, (10,2))
print(np.array([distance(P0,P1,p_i) for p_i in p]))

# 80. 配列Zに対し、与えられたpositionを配列の中心とし、与えられた形(shape)をくり抜いて新しい配列を作る(ないところは0で埋める)
# リスト内包表記
# slice
print("80:")
Z = np.random.randint(0,10,(10,10))
shape = (5,5)
fill  = 0
position = (0,0)

R = np.ones(shape, dtype=Z.dtype)*fill
P  = np.array(list(position)).astype(int)
Rs = np.array(list(R.shape)).astype(int)
Zs = np.array(list(Z.shape)).astype(int)

R_start = np.zeros((len(shape),)).astype(int)
R_stop  = np.array(list(shape)).astype(int)
Z_start = (P-Rs//2)
Z_stop  = (P+Rs//2)+Rs%2

R_start = (R_start - np.minimum(Z_start,0)).tolist()
Z_start = (np.maximum(Z_start,0)).tolist()
R_stop = np.maximum(R_start, (R_stop - np.maximum(Z_stop-Zs,0))).tolist()
Z_stop = (np.minimum(Z_stop,Zs)).tolist()

r = [slice(start,stop) for start,stop in zip(R_start,R_stop)]
z = [slice(start,stop) for start,stop in zip(Z_start,Z_stop)]
R[r] = Z[z]
print(Z)
print(R)
```
